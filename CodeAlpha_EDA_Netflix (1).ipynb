## Statistical Summary

The `describe()` function generates descriptive statistics for the numerical columns in the dataset. It provides important statistical measures such as count, mean, standard deviation, minimum value, maximum value, and quartiles. These statistics help us understand the distribution and characteristics of the numerical data.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

## Movies vs TV Shows Distribution

This visualization compares the number of Movies and TV Shows available in the Netflix dataset. It helps us understand the distribution of content types on the platform.

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar', color='green')

plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

## Top 10 Countries with Netflix Content

This visualization shows the top 10 countries that have the highest number of movies and TV shows available on Netflix. It helps identify which countries contribute the most content to the Netflix platform.

```python
plt.figure(figsize=(8,5))

df['rating'].value_counts().head(10).plot(kind='bar', color='red')

plt.title("Most Common Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

## Most Common Content Ratings

Content ratings indicate the age group for which a movie or TV show is suitable. This visualization shows the most common content ratings in the Netflix dataset, helping us understand the target audience of Netflix content.

```python
plt.figure(figsize=(12,5))

plt.hist(df['release_year'], bins=20)

plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

## Distribution of Release Years

This visualization shows how Netflix content is distributed across different release years. It helps identify trends in content production and understand which time periods have the highest number of movies and TV shows.

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Movies vs TV Shows")
plt.ylabel("")
plt.show()
```

## Movies vs TV Shows Percentage Distribution

A pie chart is used to visualize the percentage distribution of Movies and TV Shows in the Netflix dataset. It provides a clear view of the proportion of each content type.

### Observation

- Movies account for the majority of Netflix content.
- TV Shows make up a smaller percentage of the Netflix library.
- The pie chart clearly shows that Netflix offers more movies than TV Shows.
